StandardScaler => variables will follow standard normal distribution (mean 0; variance=1)


```python
df_scaled = StandardScaler().fit_transform(df_2.iloc[:,:-1])
df_scaled =pd.DataFrame(df_scaled)
df_scaled.head(5)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)

pca.fit(df_scaled)
```

```python
df_pca = pca.transform(df_scaled)

print('type: {0} and shape: {1}'.format(type(df_pca), df_pca.shape))
```

Generally the PCA is set using the option that allows you to set the variance of the input that is supposed to be explained by the generated components. The explained variance to be between 95–99%. Here the variance level I have taken as 95%. For this 15 number of components are suitable. 

```python
pca_columns = ['pca_component_1','pca_component_2','pca_component_3','pca_component_4','pca_component_5','pca_component_6','pca_component_7','pca_component_8','pca_component_9'
,'pca_component_10','pca_component_11','pca_component_12','pca_component_13','pca_component_14','pca_component_15']
df_pca = pd.DataFrame(df_pca, columns = pca_columns)
df_pca['Y'] = X_target
df_pca.head(45)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

SVD

```python
df_ts =csr_matrix(df_scaled)
```

```python
tsvd = TruncatedSVD(n_components=df_ts.shape[1]-1)
X_tsvd = tsvd.fit(df_scaled)
```

Determining the best number of components for SVD by Truncating the value. Truncated SVD generates the matrices with the specified number of columns from features. 22 is the best number of components for SVD.

```python
X_tsvd
```

```python
tsvd = TruncatedSVD(n_components=22)
tsvd.fit(df_scaled)
df_tsvd=tsvd.transform(df_scaled)
```

```python
U, Sigma, Vt = svd(df_tsvd)
print(U.shape, Sigma.shape, Vt.shape)
print('U matrix: \n', np.round(U, 2))
print('Sigma values: \n', np.round(Sigma, 2))
print('Vt matrix: \n', np.round(Vt, 2))
```
